```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
numpy.random.seed(7)

list1 = []
list2 = []

for i in range(1000):
    list1.append(numpy.random.randint(100))
    list2.append(numpy.random.randint(100))
    
   
X = numpy.column_stack([list1, list2])

list3 = []

for val in X:
    list3.append(val[0]*val[1])

Y = numpy.array(list3)




model = Sequential()
model.add(Dense(24, input_dim=2, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='relu'))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, epochs=80, batch_size=10)



scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
z = numpy.array([[11,2],[21,33]])

predictions = model.predict(z)
print(predictions)
```
